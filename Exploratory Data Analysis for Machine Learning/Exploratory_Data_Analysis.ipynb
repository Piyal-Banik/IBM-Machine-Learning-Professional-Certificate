## Data Cleaning

### Missing Value

```python
missing_data = dataset.isnull()

for column in missing_data.columns.values.tolist():
    print(column)
    print(str(np.sum(missing_data[column])) + ' missing values - ' + str(np.sum(missing_data[column]) / 145460 * 100) + " %" )
    print("")
```

```python
dataset.drop(['Evaporation', 'Sunshine', 'Cloud9am', 'Cloud3pm'], axis = 1, inplace = True)
```

```python
for column in dataset.columns.values.tolist():
    if(dataset[column].dtype == np.float64):
        mean = dataset[column].mean()
        dataset[column].replace(np.NaN, mean, inplace = True)
    else:
        most_common = dataset[column].value_counts().idxmax()
        dataset[column].replace(np.NaN, most_common, inplace = True)
```
